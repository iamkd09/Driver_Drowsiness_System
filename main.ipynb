```python
mixer.init()

# Load sound
sound = mixer.Sound('/home/lenovo/Driver_Drowiness_Detection/alarm.wav')

# Load cascade classifiers for face and eye detection
# face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
# eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Load your trained model
# Replace 'model' with the actual model you're using for prediction
# model = ...

cap = cv2.VideoCapture(0)
Score = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    height, width = frame.shape[0:2]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
    eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)
    
    cv2.rectangle(frame, (0, height-50), (200, height), (0, 0, 0), thickness=cv2.FILLED)
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, pt1=(x, y), pt2=(x+w, y+h), color=(255, 0, 0), thickness=2)
        
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(frame, pt1=(ex, ey), pt2=(ex+ew, ey+eh), color=(0, 255, 0), thickness=2)
        
        # Preprocessing steps
        eye = frame[ey:ey+eh, ex:ex+ew]
        eye = cv2.resize(eye, (80, 80))
        eye = eye / 255.0
        eye = eye.reshape(80, 80, 3)
        eye = np.expand_dims(eye, axis=0)
        
        # Model prediction (Replace 'model' with your actual model)
        prediction = model.predict(eye)
        
        # If eyes are closed
        if prediction[0][0] > 0.30:
            cv2.putText(frame, 'closed', (ex, ey-30), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,
                        fontScale=1, color=(255, 255, 255), thickness=1, lineType=cv2.LINE_AA)
            cv2.putText(frame, 'Score'+str(Score), (ex, ey-10), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,
                        fontScale=1, color=(255, 255, 255), thickness=1, lineType=cv2.LINE_AA)
            Score += 1
            if Score > 15:
                sound.play()

        # If eyes are open
        elif prediction[0][1] > 0.90:
            cv2.putText(frame, 'open', (ex, ey-30), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,
                        fontScale=1, color=(255, 255, 255), thickness=1, lineType=cv2.LINE_AA)
            cv2.putText(frame, 'Score'+str(Score), (ex, ey-10), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,
                        fontScale=1, color=(255, 255, 255), thickness=1, lineType=cv2.LINE_AA)
            Score -= 1
            if Score < 0:
                Score = 0
                sound.stop()
            
    cv2.imshow('frame', frame)
    if cv2.waitKey(33) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
